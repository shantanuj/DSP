```python
'''Convolution functions'''

'''Convolution of a signal'''

def convolve(input_signal, impulse_response):
    num_input_samples= input_signal.shape[0]
    num_response_steps = impulse_response.shape[0]
    output_convolve = np.zeros(num_input_samples+num_response_steps-1)
    
    for sample_step in range(num_input_samples):
        output_convolve[sample_step:sample_step+num_response_steps] += input_signal[sample_step] * impulse_response
        
    return output_convolve


'''Convolution when sparse response'''
def sparse_impulse_convolution(input_signal, impulse, pos_non_zeros_list):
    impulse_duration = len(impulse)
    input_signal_duration = len(input_signal)
    res_convolution = np.zeros(input_signal_duration+impulse_duration-1)
    for non_zero_pos in pos_non_zeros_list:
        res_convolution[non_zero_pos:non_zero_pos+input_signal_duration]+= impulse[non_zero_pos]*input_signal
    
    return res_convolution
```

```python
'''System class'''
class system(object):
    def __init__(self, impulse_response, convolve_func = np.convolve):
        self.impulse_response = impulse_response
        self.convolve_func = convolve_func
        
    def system_out(self, input_, custom_convolve=None, pos_non_zeros_list = None):
        if(custom_convolve is not None and pos_non_zeros_list is None):
            return custom_convolve(input_, self.impulse_response)
        elif(pos_non_zeros_list is not None):
            #Assuming non zeros list is only used with faster convolution
            return custom_convolve(input_, self.impulse_response, pos_non_zeros_list)
        else:
            return self.convolve_func(input_, self.impulse_response)
                        
```

```python
#y[0] = b[m]*x[n-m]  - a[k]*y[n-k]
def linear_filter(A,B,X):
    
    x_look_back_num = len(B)
    y_look_back_num = len(A)
    
    memX = np.zeros(x_look_back_num) #the memory storage
    memY = np.zeros(y_look_back_num)
    
    A_coeffs = np.array(A) #the constant coeffs
    B_coeffs = np.array(B)
    
    y= np.zeros(len(X)) #the output
    
    for i in np.arange(len(X)):
        memX[0] = X[i]
        memY[0] = y[i]
        feedforward = np.sum(np.multiply(memX, B_coeffs))
        #ASSUMING MEMORY<<Num_y_values
        
        feedback = np.sum(np.multiply(memY, A_coeffs))
        
        #feedback = np.sum(np.multiply(np.flip(y[i-y_look_back_num:i],-1), A_coeffs))
        y[i] = feedforward - feedback
        
        memX = np.roll(memX,1) #this is a circular shift
        memX[0] = 0 #avoiding circular shift
        
        memY[0] = y[i]
        memY = np.roll(memY,1)
         
    return y  
        
B = [1, -0.7653668, 0.99999]
A = [1, -0.722744, 0.888622]    
```
